```python
from IPython.display import clear_output
def userInput():
    validValue = ['X','O']
    player1=''
    while(player1 not in validValue):
        player1 = input("Please pick a marker X or O:").upper()
    return player1

from IPython.display import clear_output
def replay():
    replay=''
    while replay not in ['yes','no']:
        replay = input("Do you want to play again?(yes/no)")
    return replay

def player_choice(board):
    position=positionInput()
    while not space_check(board,position):
        print("No space..Please reenter")
        position = positionInput()
    return position

def full_board_check(board):
    isFull = True
    for i in board:
        isFull = isFull and i != ' '
    return isFull

def space_check(board, position):
    return board[position] == ' '

import random

def choose_first():
    return random.randint(1,2)

def win_check(board, mark):
    successpath=[[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,6,9],[4,5,6],[7,8,9],[3,5,7]]
    for path in successpath:
        win_check=True
        for index in path:
            win_check = win_check and board[index]==mark
        if win_check:
            break
    return win_check

def place_marker(board, marker, position):
    board[position]=marker

from IPython.display import clear_output

def display_board(board=[' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']):
    print(' \t| \t| \t');
    print(f'   {board[1]}    |   {board[2]}   |   {board[3]}    ');
    print(' \t| \t| \t');
    print('------------------------')
    print(' \t| \t| \t');
    print(f'   {board[4]}    |   {board[5]}   |   {board[6]}    ');
    print(' \t| \t| \t');
    print('------------------------')
    print(' \t| \t| \t');
    print(f'   {board[7]}    |   {board[8]}   |   {board[9]}    ');
    print(' \t| \t| \t');

from IPython.display import clear_output
def positionInput():
    validValue = range(1,10)
    position=0
    while(position not in validValue):
        try:
            position = int(input("Please Enter a number(1-9):"))
        except ValueError:
            position=0
    return position

def gamePlay(test_board,marker):
    position = player_choice(test_board)
    place_marker(test_board,marker,position)
    clear_output()
    display_board(test_board)
```

```python
from IPython.display import clear_output
clear_output()
while True:
    clear_output()
    print('Welcome to Tic Tac Toe!')
    test_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    display_board(test_board)
    flag = choose_first()
    print(f'player {flag} starts first')
    print(f'player {flag}:')
    marker1=userInput()
    print(f'player {1 if flag==2 else 2}:')
    marker2=userInput()
    while not full_board_check(test_board) and not win_check(test_board,marker2):
        gamePlay(test_board,marker1)
        if not full_board_check(test_board) and not win_check(test_board,marker1):
            gamePlay(test_board,marker2)
        else:
            break
    if win_check(test_board,marker1):
        print(f"Player {flag} wins")
    elif win_check(test_board,marker2):
        print(f"Player {1 if flag==2 else 2} wins")
    else:
        print("It is a draw!")
    replayFlag = replay()
    if replayFlag == 'no':
        break;
```
